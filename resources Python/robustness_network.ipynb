```python
import matplotlib.pyplot as plt
import networkx as nx
import random

# these are for the drawing function
g = nx.Graph().to_undirected()
node_pose = {}
drawGraphOn = True
plotOn = True
diametersOn = True
plotDiametersOn = False

def remove_random_node(g, n):
    import random  # This because it crashes against the import at the top which works for the plot
    for i in range(n):
        node = random.choice(list(g.nodes.keys()))
        g.remove_node(node)
    return g    


def draw_graph(g):
    if g.number_of_nodes() < 100:
        nx.draw_networkx(g, node_size=150, font_size=9, arrowsize=3, node_color='b')
    else:
        nx.draw_networkx(g, node_size=50, font_size=6, arrowsize=3, node_color='b')
    
```

# An example

```python
# ER graph (with 100 nodes and probability to create an edge p=0.5)
G = nx.erdos_renyi_graph(100, 0.02)
```

```python
#graph before random attack
draw_graph(G)
```

```python
#graph after random attack
n = 30
G1 = remove_random_node(G, n)
draw_graph(G1)
```
